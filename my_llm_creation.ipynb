```python
# -*- coding: utf-8 -*-
"""
Kalkulator Sederhana untuk Google Colab dengan Output Berwarna Pink (Magenta).
"""

# Kode ANSI untuk warna
PINK = '\033[95m'  # Kode untuk warna magenta cerah (mirip pink)
RESET = '\033[0m'  # Kode untuk mereset warna kembali ke default
BIRU = '\033[94m'   # Kode untuk warna biru (untuk prompt)

def tambah(x, y):
  """Fungsi untuk menjumlahkan dua angka."""
  return x + y

def kurang(x, y):
  """Fungsi untuk mengurangkan dua angka."""
  return x - y

def kali(x, y):
  """Fungsi untuk mengalikan dua angka."""
  return x * y

def bagi(x, y):
  """Fungsi untuk membagi dua angka. Menangani pembagian dengan nol."""
  if y == 0:
    return "Error: Tidak bisa membagi dengan nol!"
  return x / y

def kalkulator():
  """Fungsi utama untuk menjalankan kalkulator."""
  print(f"{PINK}======================================{RESET}")
  print(f"{PINK}   Kalkulator Sederhana Nuansa Pink   {RESET}")
  print(f"{PINK}======================================{RESET}")
  print("Masukkan 'quit' kapan saja untuk keluar.\n")

  while True:
    try:
      # Meminta input angka pertama
      angka1_input = input(f"{BIRU}Masukkan angka pertama: {RESET}")
      if angka1_input.lower() == 'quit':
        break
      angka1 = float(angka1_input)

      # Meminta input operator
      operator = input(f"{BIRU}Masukkan operator (+, -, *, /): {RESET}")
      if operator.lower() == 'quit':
        break
      # Validasi operator
      if operator not in ['+', '-', '*', '/']:
          print(f"{PINK}Operator tidak valid. Coba lagi.{RESET}")
          continue

      # Meminta input angka kedua
      angka2_input = input(f"{BIRU}Masukkan angka kedua: {RESET}")
      if angka2_input.lower() == 'quit':
        break
      angka2 = float(angka2_input)

      # Melakukan perhitungan berdasarkan operator
      hasil = None
      if operator == '+':
        hasil = tambah(angka1, angka2)
      elif operator == '-':
        hasil = kurang(angka1, angka2)
      elif operator == '*':
        hasil = kali(angka1, angka2)
      elif operator == '/':
        hasil = bagi(angka1, angka2)

      # Mencetak hasil dengan warna pink
      print(f"{PINK}--------------------------------------{RESET}")
      if isinstance(hasil, str): # Jika hasilnya adalah pesan error
          print(f"{PINK}Hasil: {hasil}{RESET}")
      else:
          print(f"{PINK}Hasil: {angka1} {operator} {angka2} = {hasil}{RESET}")
      print(f"{PINK}--------------------------------------{RESET}\n")

    except ValueError:
      # Menangani jika input bukan angka yang valid
      print(f"{PINK}Input tidak valid. Masukkan angka yang benar.{RESET}\n")
    except Exception as e:
      # Menangani error tak terduga lainnya
      print(f"{PINK}Terjadi error: {e}{RESET}\n")

  print(f"\n{PINK}Terima kasih telah menggunakan kalkulator!{RESET}")

# Menjalankan fungsi kalkulator
if __name__ == "__main__":
    kalkulator()
```
